```python
searchQuery = '''
                    (climate change) OR #climatechange OR 
                    (global warming) OR #globalwarming OR 
                    (climate crisis) OR #climatecrisis OR
                    (climate alarmist) OR (climate alarmists) OR
                    (climate denier) OR (climate deniers) OR
                    (climate skeptic) OR (climate skeptics) OR 
                    (climate skepticist) OR (climate skepticists)
                                                                    '''
```

```python
sinceId = None
max_id = None
maxTweets = 2000000
tweetsPerQry = 100
tweetCount = 0
tweetFiles = 0
tweets = []
searchCount = 0

while 10000*tweetFiles + tweetCount < maxTweets:
    try:
        if (not max_id):
            if (not sinceId):
                new_tweets = api.search_tweets(q=searchQuery, 
                                               tweet_mode='extended',
                                               count=tweetsPerQry)
        
            else:
                new_tweets = api.search_tweets(q=searchQuery, 
                                               tweet_mode='extended',
                                               count = tweetsPerQry, 
                                               since_id = sinceId)

        else: 
            if (not sinceId):
                new_tweets = api.search_tweets(q=searchQuery, 
                                               tweet_mode='extended',
                                               count= tweetsPerQry, 
                                               max_id=str(max_id -1))

            else:
                new_tweets = api.search_tweets(q=searchQuery,
                                               tweet_mode='extended',
                                               count=tweetsPerQry, 
                                               max_id=str(max_id -1), 
                                               since_id=sinceId)    

        new_tweets = new_tweets['statuses']
            
        if len(new_tweets) == 0:
            msg = 'No remaining tweets found'
            logging.error(msg)
            break
        
        tweets.extend(new_tweets)
        tweetCount += len(new_tweets)
        msg = f'{tweetCount} tweets downloaded'
        logging.error(msg)
        max_id = new_tweets[-1]['id']
    
        if tweetCount >= 10000:
            with open('./tweets'+str(tweetFiles)+'.json', 'w') as outfile:
                json.dump(tweets, outfile)
            msg = 'JSON file saved'
            logging.error(msg)
            tweetFiles += 1
            tweetCount = 0
            del tweets[:]
        
    except tweepy.errors.TweepyException as e:
        msg = f'Query failed at max_id {max_id} : {e}'
        logging.error(msg)

if tweetCount > 0:
    with open('./tweets'+str(tweetFiles)+'.json', 'w') as outfile:
        json.dump(tweets, outfile)
    msg = 'JSON file saved'
    logging.error(msg)
    tweetFiles += 1
    tweetCount = 0
    del tweets[:]
```
